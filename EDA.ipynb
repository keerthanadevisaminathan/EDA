# Z-score -> Z=x-μ/σ

```python
from scipy import stats
zscore=np.abs(stats.zscore(data['normalized-losses']))
zscore
```

```python
threshold=zscore.max()
threshold
index_val=np.where(zscore>=threshold)
data.iloc[index_val]['normalized-losses']
```

```python
new_data=data.drop(index=data.iloc[index_val]['normalized-losses'].index)
new_data
```
